```python
import streamlit as st
import requests

# Set title for the app
st.title("Car Price Prediction")

# Create input fields for the required features
brand = st.text_input("Brand")
price = st.number_input("Price", min_value=0)
body = st.text_input("Body")
mileage = st.number_input("Mileage", min_value=0)
engine_v = st.number_input("Engine Volume", min_value=0.0)
engine_type = st.text_input("Engine Type")
registration = st.text_input("Registration")
year = st.number_input("Year", min_value=1900, max_value=2024)
model = st.text_input("Model")

# When the user clicks on the "Predict" button
if st.button("Predict"):
    # Prepare the data for sending to FastAPI
    data = {
        "Brand": brand,
        "Price": price,
        "Body": body,
        "Mileage": mileage,
        "EngineV": engine_v,
        "Engine_Type": engine_type,
        "Registration": registration,
        "Year": year,
        "Model": model
    }
    
    # Send the data to the FastAPI endpoint
    response = requests.post("http://127.0.0.1:8000/predict", json=data)
    
    # Display the result
    if response.status_code == 200:
        prediction = response.json()
        st.success(f"The predicted price is: {prediction['prediction']}")
    else:
        st.error("Error in prediction")

```

```python
streamlit run app.py
```
